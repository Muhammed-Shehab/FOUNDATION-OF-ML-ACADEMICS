Task 2: Calculate average electicity usage of every hostel

```python
avg = electricity.mean(axis=1)
print(avg)
```

Task C1: Create a new array penalized usage such that:
-> values > 550 -> multiplied by 1.20
-> all other values remain unchanged

```python
penalized = np.where(electricity > 550, electricity * 1.20, electricity)
print(penalized)
```

Task D1: compare each hostel's daily usage with its own weekly average . Create a boolean matrix where:
->True : above hostel average
->False : below or equal to hostel average

```python
boolean = electricity > avg[:, np.newaxis]
print(boolean)
```

Task E1: reshape the electricity data into a single row representing weekly usage (shape : 1x28)

```python
electricity = electricity.reshape(1, 28)
print(electricity)
```

Task E2: reshape the same data into 2 rows and 14 columns. Ensure no data is lost or duplicated

```python
electricity = electricity.reshape(2, 14)
print(electricity)
```

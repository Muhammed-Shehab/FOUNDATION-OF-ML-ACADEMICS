```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Name": ["Aman", "Saurav", "Karan", "Shubham", "Abhishek", None, "Sameer"],
    "Age": [22, 23, None, 22, 25, 24, 26],
    "Marks": [85, None, 88, 90, 56, 72, None],
    "Subject": ["Math", "Physics", "Math", "Chemistry", "Math", "Physics", "Chemistry"]
}
df = pd.DataFrame(data);
print(df)
```

### basic groupby

```python
df.groupby('Subject')['Marks'].mean()                      # mean marks per subject
```

```python
df.groupby('Subject')['Name'].count()                    # Count student per subject
```

```python
df.groupby('Age')['Name'].count()
```

```python
df.groupby('Subject').size()                              # this will include the missing values in the count also
```

#### Multiple aggreagation

```python
df.groupby('Subject')['Marks'].agg(['mean', 'max', 'min'])
```

#### Group by Multiple Columns

```python
df.groupby(['Subject', 'Age'])['Marks'].mean()
```

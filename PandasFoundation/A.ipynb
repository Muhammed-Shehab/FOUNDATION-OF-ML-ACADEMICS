DATAFRAME IN PANDAS

```python
data = {
    "Name": ["Aman", "Krish", "Karan"],
    "Ages": [22, 23, 21],
    "Marks": [85, 90, 88]
}
print(data)
type(data)
```

```python
marks = pd.DataFrame(data)
print(marks)                    
#Pandas automatically adds default index: This is called a RangeIndex (starts from 0 by default).
```

```python
#We can also add cutom index in pandas dataframe
marks = pd.DataFrame(
    data,
    index = ["Student1", "Student2", "Student3"]
)
print(marks)
```

```python
marks.set_index("Name", inplace=True) #if we are setting inplace = true it means we changing the original object itself, if we don't do an another is
print(marks)
```

HANDLING MISSING VALUES

```python
#updating the dataset will null values

df = pd.DataFrame({
    "Name": ["Aman", "Saurav", "Karan", "Shubham", None, "Sameer"],
    "Age": [22, 23, None, 22, 25, 26],
    "Marks": [85, None, 88, 90, 56, None]
})
df
```

```python
df.isnull()          #checking the missing values
```

```python
df.isnull().sum()                            #counting the missing values
```

```python
df.dropna()                          #it will drop the whole row which has missing values
```

```python
print(df)                            #unless we use inplace = true the dataframe remain the same as it is
cleaned_data = df.dropna()           #so we can create a new dataframe
cleaned_data
```

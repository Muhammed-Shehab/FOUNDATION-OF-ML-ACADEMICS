```python
import pandas as pd
import numpy as np
```

```python
data = {
    "Name": ["Aman", "Saurav", "Karan", "Shubham","Abhishek", None, "Sameer"],
    "Age": [22, 23, None, 22, 25, 24, 26],
    "Marks": [85, None, 88, 90, 56, 72, None],
    "Subject" : ["Math", "Physics", "Math", "Chemistry", "Math", "Physics", "Chemistry"]
}
df = pd.DataFrame(data)
```

```python
print(df.shape)
print(df.dtypes)
print(df.describe().T)
```

```python
df[["Name", "Marks"]]
```

### filering - display students with more than 80 marks

```python
df[df["Marks"] > 80]
```

#### Complex filtering - display students with more than 80 marks and older than 22 years

```python
df[(df["Marks"] > 80) & (df["Age"] > 22)]
```

#### Aggregation - count how many student are enrolled in each subject

```python
df["Subject"].value_counts()
```

#### Sort the dataframe by 'Marks' in descending order

```python
df["Marks"].sort_values(ascending=False)
```

#### Use a specific Pandas function to find the top 2 highest scores (Scholarship Candidates) without sorting the whole dataframe manually

```python
df.nlargest(2, "Marks")
```

#### Identify the student with the lowest marks(Remedial Candidate)

```python
df.loc[df["Marks"].idxmin()]
```

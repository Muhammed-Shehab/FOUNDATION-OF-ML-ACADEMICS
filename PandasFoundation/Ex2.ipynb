```python
import numpy as np
import pandas as pd
```

```python
data = {
    "Name": ["Aman", "Sagar", "Shan", "Arip", None, "Anzil", None, "Saneej", "Rafi", "Nick"],
    "Department": ['IT', 'Stock','Stock','IT','Stock', 'IT','HR','HR','HR', 'IT'],
    "Role": ['Dev', 'Quant', 'Quant', 'Cyber', 'Quant', 'Dev', 'Sales', 'Sales', 'Sales', 'Cyber'],
    "Experience": [4, 6, 5, 8, 2, 3, 4, 4, 9, 7],
    "Salary": [40000,100000,100000,35000,80000,40000,40000,55000,60000,35000],
    "Score": [6, 8, 10, 9, 8, 7, 7, 5, 9, 6]
}
df = pd.DataFrame(data)
df
```

What is the average salary of each Department?

```python
df.groupby('Department')['Salary'].mean()
```

What is the average perfomance score for each role?

```python
df.groupby("Role")["Score"].mean()
```

How many actual, non - null perfomance scores entries exist for each department?

```python
df.groupby("Department")["Score"].count()
```

What is the total number of employee recordes(rows) in each department, regardless of whether data is missing?

```python
df["Department"].value_counts()
```

Generate a summary table for Salary based on department that show the mean, maximum, minimum, and count in a single line of code.

```python
df.groupby("Department")["Salary"].agg(["mean", "max", "min", "count"])
```

Generate the same summary table (mean, max, min, count) but analyze the perfomance score based on the role.

```python
df.groupby("Role")["Score"].agg(["mean", "max", "min", "count"])
```

Find the average salary grouped by both department and role.

```python
df.groupby(["Department", "Role"])["Salary"].mean()
```

Find the average perfomance score grouped by both department and experience level

```python
df.groupby(["Department", "Experience"])["Score"].mean()
```
